```python
df = df.dropna(subset=["co2e_estimated_kg"])
df.fillna(0, inplace=True)
```

```python
df["hour"] = df["timestamp"].dt.hour
df["dow"] = df["timestamp"].dt.dayofweek
df["is_weekend"] = (df["dow"] >= 5).astype(int)
```

```python
df.shape
```

```python
# 0) Ensure a site key exists so we can group per site (or create a single-site fallback)
if "site_id" not in df.columns:
    df["site_id"] = "SITE_1"

# 1) Standardize expected driver/target column names (only when present)
rename_map = {
    "energy_usage_kwh": "electricity_kwh",
    "fuel_liters": "diesel_litres",
    "temperature_c": "temp_c",
    "equipment_load_%": "equipment_load_pct",
    "shift_id": "shift",
    "co2e_estimated_kg": "co2e_observed_kg",
}
df = df.rename(columns={k:v for k,v in rename_map.items() if k in df.columns})

# 2) Ensure numeric types for driver columns we will use
for c in ["electricity_kwh","diesel_litres","output_tons","temp_c","equipment_load_pct"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

# 3) Sort for time-aware features
df = df.sort_values(["site_id","timestamp"])

# 4) Build lags & rolling stats safely (works even if only a single site)
drivers = [c for c in ["electricity_kwh","diesel_litres","output_tons"] if c in df.columns]

for col in drivers:
    # Grouped Series (per site)
    g = df.groupby("site_id")[col]

    # Lags
    df[f"{col}_lag1"] = g.shift(1)
    df[f"{col}_lag4"] = g.shift(4)  # ~1 hour back if 15-min data

    # Rolling means (reset index to align back to original index)
    df[f"{col}_roll4"]  = (
        g.rolling(4, min_periods=1).mean()
         .reset_index(level=0, drop=True)
    )
    df[f"{col}_roll96"] = (
        g.rolling(96, min_periods=1).mean()
         .reset_index(level=0, drop=True)
    )

# 5) Fill NaNs that were created by shift/rolling at the series start
eng_cols = [c for c in df.columns if any(s in c for s in ["_lag1","_lag4","_roll4","_roll96"])]
df[eng_cols] = df[eng_cols].fillna(0)
```

```python
print("Added engineered columns:", eng_cols)
print("Sample:")
print(df.head(5)[["timestamp","site_id"] + drivers + eng_cols[:6] + ["hour","dow","is_weekend"] if drivers else ["timestamp","site_id","hour","dow","is_weekend"]])
```

```python
# ===== 6) Select features & target =====
target = "co2e_observed_kg" if "co2e_observed_kg" in df.columns else "co2e_estimated_kg"

base_feats = [c for c in [
    "electricity_kwh","diesel_litres","output_tons",
    "temp_c","equipment_load_pct","shift",
    "hour","dow","is_weekend"
] if c in df.columns]

eng_feats = [c for c in df.columns if c.endswith(("_lag1","_lag4","_roll4","_roll96"))]
features  = base_feats + eng_feats
print("Using features:", len(features), "→", features[:10], "..." if len(features) > 10 else "")

X = df[features].copy()
y = df[target].astype(float).copy()

# safety: replace inf/NaN that might appear after engineering
import numpy as np
X = X.replace([np.inf, -np.inf], np.nan).fillna(0)

# ===== 7) Chronological split (no shuffle) =====
split_idx = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# ===== 8) Preprocess + train model =====
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt

categorical = [c for c in ["shift"] if c in X.columns]
numeric     = [c for c in X.columns if c not in categorical]

# Add StandardScaler for numeric features
if len(categorical) > 0:
    preprocess = ColumnTransformer([
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical),
        ("num", StandardScaler(), numeric), # Added StandardScaler here
    ])
else:
    # no categorical columns; just pass numeric through
    preprocess = StandardScaler() # Added StandardScaler here

model = Pipeline([
    ("prep", preprocess),
    ("reg", RandomForestRegressor(
        n_estimators=400, max_depth=None, random_state=42, n_jobs=-1
    ))
])

model.fit(X_train, y_train)

# ===== 9) Evaluate (compat with older sklearn) =====
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = sqrt(mean_squared_error(y_test, y_pred))   # <— no squared=False
r2 = r2_score(y_test, y_pred)
print(f"Test MAE:  {mae:,.2f}")
print(f"Test RMSE: {rmse:,.2f}")
print(f"Test R²:   {r2:,.3f}")

# ===== 10) Save model =====
import joblib, json
joblib.dump({"model": model, "features": features}, "emissions_model.joblib")
with open("model_manifest.json","w") as f:
    json.dump({
        "features": features,
        "target": target,
        "rows": int(len(df)),
        "train_rows": int(len(X_train)),
        "test_rows": int(len(X_test)),
        "metrics": {"mae": float(mae), "rmse": float(rmse), "r2": float(r2)}
    }, f, indent=2)
print("Saved emissions_model.joblib and model_manifest.json")

# ===== 11) (Optional) simple prediction helper =====
def predict_emissions(sample: dict):
    """
    sample keys must match the training feature names.
    Returns predicted total kg CO2e and simple intensities if denominators provided.
    """
    import pandas as pd, joblib
    pack = joblib.load("emissions_model.joblib")
    mdl, feats = pack["model"], pack["features"]
    Xnew = pd.DataFrame([sample])[feats].replace([np.inf, -np.inf], np.nan).fillna(0)
    yhat = float(mdl.predict(Xnew)[0])
    intensity = {}
    if "electricity_kwh" in Xnew.columns and Xnew.at[0, "electricity_kwh"] > 0:
        intensity["kg_per_kwh"] = yhat / Xnew.at[0, "electricity_kwh"]
    if "output_tons" in Xnew.columns and Xnew.at[0, "output_tons"] > 0:
        intensity["kg_per_ton"] = yhat / Xnew.at[0, "output_tons"]
    return {"co2e_total_kg": yhat, "intensity": intensity}
```

```python
from math import isfinite

# 1) Assemble a tidy evaluation frame for the test window
eval_df = df.iloc[split_idx:].copy()
eval_df = eval_df.reset_index(drop=True)

eval_df["y_true"] = y_test.values
eval_df["y_pred"] = y_pred

# 2) Carbon intensity metrics (safe divide: ignore zero denominators)
def safe_div(num, den):
    den = np.where((den is None) | (np.asarray(den)==0), np.nan, den)
    return np.divide(num, den)

if "electricity_kwh" in eval_df.columns:
    eval_df["kg_per_kwh_true"] = safe_div(eval_df["y_true"], eval_df["electricity_kwh"])
    eval_df["kg_per_kwh_pred"] = safe_div(eval_df["y_pred"], eval_df["electricity_kwh"])

if "output_tons" in eval_df.columns:
    eval_df["kg_per_ton_true"] = safe_div(eval_df["y_true"], eval_df["output_tons"])
    eval_df["kg_per_ton_pred"] = safe_div(eval_df["y_pred"], eval_df["output_tons"])

# 3) Error metrics per row
eval_df["abs_err"] = np.abs(eval_df["y_true"] - eval_df["y_pred"])
eval_df["pct_err"] = safe_div(eval_df["y_pred"] - eval_df["y_true"], eval_df["y_true"]) * 100.0

# 4) Summary table
def mape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    mask = y_true != 0
    return np.mean(np.abs((y_pred[mask] - y_true[mask]) / y_true[mask])) * 100.0

summary = {
    "rows": int(len(eval_df)),
    "MAE": float(eval_df["abs_err"].mean()),
    "RMSE": float(np.sqrt(np.mean((eval_df["y_pred"] - eval_df["y_true"])**2))),
    "R2_recomputed": float(1 - np.sum((eval_df["y_true"]-eval_df["y_pred"])**2) / np.sum((eval_df["y_true"]-eval_df["y_true"].mean())**2 + 1e-12)),
    "MAPE_%": float(mape(eval_df["y_true"], eval_df["y_pred"]))
}
print("=== Test Summary ===")
for k,v in summary.items():
    if isinstance(v, float):
        print(f"{k}: {v:,.4f}")
    else:
        print(f"{k}: {v}")

# 5) Daily averages (nicer for dashboards)
eval_df["date"] = pd.to_datetime(eval_df["timestamp"]).dt.normalize()

daily_cols = ["y_true","y_pred"]
if "kg_per_kwh_true" in eval_df: daily_cols += ["kg_per_kwh_true","kg_per_kwh_pred"]
if "kg_per_ton_true" in eval_df: daily_cols += ["kg_per_ton_true","kg_per_ton_pred"]

daily = eval_df.groupby("date", as_index=False)[daily_cols].mean()

print("\n=== Head of intensity report (row-level) ===")
keep_cols = ["timestamp","y_true","y_pred"]
if "kg_per_kwh_true" in eval_df: keep_cols += ["kg_per_kwh_true","kg_per_kwh_pred"]
if "kg_per_ton_true" in eval_df: keep_cols += ["kg_per_ton_true","kg_per_ton_pred"]
keep_cols += ["abs_err","pct_err"]
print(eval_df[keep_cols].head(10))

print("\n=== Head of daily averages ===")
print(daily.head(10))

# 6) Simple plots (totals + intensities over time)
# Totals
plt.figure(figsize=(9,4))
plt.plot(eval_df["timestamp"], eval_df["y_true"], label="True")
plt.plot(eval_df["timestamp"], eval_df["y_pred"], label="Pred")
plt.title("Emissions (kg CO2e) — Test Window")
plt.xlabel("Time")
plt.ylabel("kg CO2e")
plt.legend()
plt.tight_layout()
plt.show()

# Intensity: kg per kWh (if available)
if "kg_per_kwh_true" in eval_df:
    plt.figure(figsize=(9,4))
    plt.plot(eval_df["timestamp"], eval_df["kg_per_kwh_true"], label="True intensity (kg/kWh)")
    plt.plot(eval_df["timestamp"], eval_df["kg_per_kwh_pred"], label="Pred intensity (kg/kWh)")
    plt.title("Carbon Intensity per kWh — Test Window")
    plt.xlabel("Time")
    plt.ylabel("kg CO2e per kWh")
    plt.legend()
    plt.tight_layout()
    plt.show()

# Intensity: kg per ton (if available)
if "kg_per_ton_true" in eval_df:
    plt.figure(figsize=(9,4))
    plt.plot(eval_df["timestamp"], eval_df["kg_per_ton_true"], label="True intensity (kg/ton)")
    plt.plot(eval_df["timestamp"], eval_df["kg_per_ton_pred"], label="Pred intensity (kg/ton)")
    plt.title("Carbon Intensity per Ton — Test Window")
    plt.xlabel("Time")
    plt.ylabel("kg CO2e per ton")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
